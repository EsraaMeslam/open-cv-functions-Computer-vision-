```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Add 2 images

```python

img_1 = cv2.imread('cat.webp')
img_2 = cv2.imread('chameleon.webp')

# Display image shapes and data types
print("Image 1 shape:", img_1.shape)
print("Image 2 shape:", img_2.shape)
print("Image 1 dtype:", img_1.dtype)
print("Image 2 dtype:", img_2.dtype)

plt.figure(figsize=(20, 10)) 

plt.subplot(1, 2, 1)
plt.imshow(img_1,cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_2,cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.show()
```

# Addweighted 2 images

```python
#the 2 images should has the same size to add them.

img_1=cv2.resize(img_1,(1024,1024))
img_2=cv2.resize(img_2,(1024,1024))

add_2_imgs=cv2.add(img_1,img_2)

plt.figure(figsize=(20, 10)) 
plt.subplot(1, 3, 1)
plt.imshow(img_1,cmap='gray')
plt.title('Image 1  ')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_2,cmap='gray')
plt.title('Image 2  ')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(add_2_imgs,cmap='gray')
plt.title('add 2 images ')
plt.axis('off')

plt.show()
```

```python
#the 2 images should has the same size to add them
img_1=cv2.resize(img_1,(1024,1024))
img_2=cv2.resize(img_2,(1024,1024))

add_2_imgs_w=cv2.addWeighted(img_1,.7,img_2,.3,10) #(img1,its opacity,img2,its opacity,gamma)# 0 for darker . more bigher==more brighter 

plt.figure(figsize=(20, 10)) 
plt.subplot(1, 3, 1)
plt.imshow(img_1,cmap='gray')
plt.title('Image 1  ')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_2,cmap='gray')
plt.title('Image 2  ')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(add_2_imgs_w,cmap='gray')
plt.title('add 2 Weighted images ')

plt.axis('off')
plt.show()
```
